```python
def make_corpus(input_path, output_file_path):
    i = 0
    file_list = glob.glob(input_path + '/*')
    stopwords_text = open('./stopwords.txt', mode='r', encoding="utf_8_sig")
    stopwords = stopwords_text.read().split('\n')
    stopwords_text.close()
    with open(output_file_path, 'w', encoding="utf_8_sig") as output_file:
        for file in file_list:
            if i % 10 == 0:
                print(i)
            i+=1
            try:
                with open(file, 'r', encoding="utf_8_sig") as input_file:
                    text = input_file.read()
                    text_nlp = nlp(text)
                    result_text = ""
                    for sent in text_nlp.sentences:
                        for word in sent.words:
                            if word.lemma not in stopwords:
                                result_text += word.lemma
                                result_text += " "
                    output_file.write(result_text.replace('\n', ' '))
                    output_file.write('\n')
            except:
                continue
```

```python
make_corpus('./Corpus', './Corpus_Processed/result.txt')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def make_table_and_dict(corpus_path, min_df, max_df, token_pattern = None, use_idf = True):
    with open(corpus_path, 'r', encoding="utf_8_sig") as corpus_file:
        vectorizer = TfidfVectorizer()
        data_vectorized = vectorizer.fit_transform(corpus_file)
    return data_vectorized, vectorizer.get_feature_names(), vectorizer.idf_
```

```python
hindi_data_vectorized, hindi_dictionary, idfs = make_table_and_dict('./Corpus_Processed/result.txt', 3, 0.8)
pairs = dict(zip(hindi_dictionary, idfs))
with open('./Corpus_Processed/cutted_hindi_dict.txt', 'w', encoding="utf_8_sig") as output_file:
    for word in hindi_dictionary:
        output_file.write(word)
        output_file.write('\n')
```

```python
def create_table(data_vectorized, k, name):
    u, sigma, vt = svds(data_vectorized, k)
    with open('./Corpus_Processed/SVD_U.npy', 'wb') as f:
        np.save(f, u)
    with open('./Corpus_Processed/SVD_sigma.npy', 'wb') as f:
        np.save(f, sigma)
    print(sigma)
    with open('./Corpus_Processed/SVD_VT.npy', 'wb') as f:
        np.save(f, vt)
    with open('./Corpus_Processed/' + name + str(k) + '.npy', 'wb') as f:
        np.save(f, np.dot(np.diag(sigma), vt).T)
```
